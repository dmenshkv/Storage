```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame with semicolon delimiter
data = pd.read_csv("Data.csv", delimiter=';')

# Convert the columns with numbers from string to float (replace commas with periods for correct decimal format)
data['My Solution'] = data['My Solution'].str.replace(',', '.').astype(float)
data['Other solution'] = data['Other solution'].str.replace(',', '.').astype(float)
data['Microsoft solution'] = data['Microsoft solution'].str.replace(',', '.').astype(float)

# Ensure 'Test Number' column is treated as string
data['Test Number'] = data['Test Number'].astype(str)

# Plotting the line chart
plt.figure(figsize=(30, 10))  # Adjusted figure size for better readability

# Plot the data with updated labels
plt.plot(data['Test Number'], data['My Solution'], label='My Solution', marker='o', linestyle='-', color='blue')
plt.plot(data['Test Number'], data['Other solution'], label='Other Solution', marker='o', linestyle='-', color='green')
plt.plot(data['Test Number'], data['Microsoft solution'], label='Microsoft Solution', marker='o', linestyle='-', color='red')

# Reduce x-axis ticks by showing every Test Number
plt.xticks(data['Test Number'][::5], fontsize=14)  # Adjusted font size for x-axis ticks

# Set custom y-axis ticks based on min/max values
y_min = int(min(data[['My Solution', 'Other solution', 'Microsoft solution']].min()))  # Min value for y-axis
y_max = int(max(data[['My Solution', 'Other solution', 'Microsoft solution']].max()))  # Max value for y-axis

# Define y-ticks with a step of 80
y_ticks = range(y_min, y_max + 1, 80)
plt.yticks(y_ticks, fontsize=14)  # Adjusted font size for y-axis ticks

# Adding labels and title with bigger font sizes
plt.xlabel('Test Number', fontsize=16)
plt.ylabel('Request Duration (ms)', fontsize=16)
plt.title('HTTP Request Duration Comparison', fontsize=20)

# Adding a legend to distinguish the lines and moving it to the upper-left corner
plt.legend(fontsize=14, loc='upper left')

# Display a grid for better readability
plt.grid(True)

# Adjust layout to avoid overlapping labels
plt.tight_layout()

# Display the chart
plt.show()
```
